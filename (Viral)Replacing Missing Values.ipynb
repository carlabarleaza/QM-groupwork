# Importing libraries and loading CSV file

```python
#libraries
import pandas as pd
import numpy as np
```

```python
# Load viral video data from the csv file
filename = 'viral_meta.csv'
df1 = pd.read_csv(filename)
```

```python
df1.head()
```

# Creating New 'Ratio' Column
### The "Ratio" Column will represent the like to dislike ratio as a percentage. It will be calculated using the following formula: likes / (likes + dislikes) * 100

```python
df1['ratio'] = df1['num_likes'] / (df1['num_likes'] + df1['num_dislikes']) * 100
```

```python
df1.head()
```

## Checking for missing values and replacing them

```python
#checking missing values
df1.info()
```

### The 'ratio' column we have created has 2 missing values!
### Let's identify the rows with missing values:

```python
#using .isna()
ratio_missing = df1['ratio'].isna()
df1.loc[ratio_missing, : ]
```

## We can now understand that these rows have missing values because the likes and dislikes are equal to 0. This creates a NaN value for the 'ratio' column, since  you cannot divide a number by 0!

## Replacing the values:

### Since the likes and dislikes for both of these videos are 0, we have to make the ratio 0

```python
#replacing the missing values with '0'
df1.loc[ratio_missing, : ] = 0
```

```python
#making sure they have been replaced!
df1.info()
```

### We no longer have missing values!
