## How are views, likes, dislikes, comment count, and other attributes correlated with each other? 

Correlation Heatmap (version 2) 
Coded following the steps from https://towardsdatascience.com/annotated-heatmaps-in-5-simple-steps-cc2a0660a27d but also indications from Szabo, B. (2020). How to Create a Seaborn Correlation Heatmap in Python? [online] Medium. Available at: https://medium.com/@szabo.bibor/how-to-create-a-seaborn-correlation-heatmap-in-python-834c0686b88e [Accessed 8 Dec. 2020].

```python
# We create the correlation matrix with .corr
corr_matrix = df.corr()
```

```python
corr_matrix
```

```python
#For greater clarity, we cut the heatmap in half to only have the lower triangle. 
#We use the seaborn heatmap ‘mask’ argument  to cover part of the heatmap.
mask = np.zeros_like(corr_matrix, dtype=np.bool) #takes a boolean array or a dataframe as an argument; when defined, cells become invisible for values where the mask is True
mask[np.triu_indices_from(mask)]= True  # isolates the upper triangle of a matrix while turning all the values in the lower triangle into 0
```

```python
f, ax = plt.subplots(figsize=(11, 15)) # Sets the size of the heatmap 
heatmap = sns.heatmap(corr_matrix, 
                      mask = mask,
                      square = True,
                      linewidths = .5,
                      cbar_kws = {'shrink': .4, 
                                'ticks' : [-1, -.5, 0, 0.5, 1]},
                      vmin = -1, # vmin and vmax set the range of values that serve as the basis for the colormap
                      vmax = 1, 
                      annot = True, # set to True makes the correlation values become visible on the colored cells
                      annot_kws = {'size': 12})
#add the column names as labels
ax.set_yticklabels(corr_matrix.columns, rotation = 0)
ax.set_xticklabels(corr_matrix.columns)
sns.set_style({'xtick.bottom': True}, {'ytick.left': True})
ax.set_title("Correlation heatmap for trending videos", fontsize=20) #This line of code was done using Stack overflow (https://stackoverflow.com/questions/32723798/how-do-i-add-a-title-to-seaborn-heatmap)
```
