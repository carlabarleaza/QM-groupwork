## Getting rid of columns we don't need (we can always decide on that later):

```python
df = df.drop(['video_id','avg_rate','max_rate','min_rate','num_raters','uploader_url','uploader_id','uploader_name'], axis=1)
```

## indentifying at missing values:

```python
#missing values are only in our video_desp column
video_desp_missing = df['video_desp'].isna()
```

```python
#lets look at the rows where we have missing values
df.loc[video_desp_missing, : ]
```

### video description is not a column we are going to analyse (so doesn't matter that we have NaN values), but I included this to show how to look at missing values!

## removing or replacing missing data:

```python
#we could fill in the missing values (again, not that important for our dataset but its good practice)
df.loc[video_desp_missing, 'video_desp'] = "description missing"
```

```python
#we can see that we don't have any missing values anymore!
df.info()
```
