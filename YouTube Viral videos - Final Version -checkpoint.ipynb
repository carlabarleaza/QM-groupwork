## Creating a date and a time column 

We want to create two separate columns indicating the date the video has been uploaded and the time. Currently, the upload_time column displays the values in the form 2008-12-28T21:57:55.000Z. This coresponds to a video published on 2008-12-28 at 21:57:55 in the Coordinated Universal Time (UTC) timezone.  

```python
df["publishing_day"] = df["upload_time"].apply(
    lambda x: datetime.datetime.strptime(x[:10], "%Y-%m-%d").date().strftime('%a'))
df["publishing_hour"] = df["upload_time"].apply(lambda x: x[11:13])
#df.drop(labels='upload_time', axis=1, inplace=True) This line of code deletes the original publish_time column, remove # to activate
```

## At which time of the day are most viral videos published?

```python
plt.figure(figsize = (14,9))
PLOT_COLORS = ["#268bd2", "#0052CC", "#FF5722", "#b58900", "#003f5c"] #this line gives different options for color palette
                  
plt.figure(figsize = (14,9))
gr = sns.countplot('publishing_hour', data=df, color=PLOT_COLORS[0])
gr.set_xticklabels(gr.get_xticklabels(),rotation=45)
gr.set_title("Number of viral videos published by hour", fontsize=20)
gr.set_xlabel("Publishing Hour (UTC timezone)", fontsize=12)
gr.set_ylabel("Number of videos", fontsize=12)
```

Interesting results... a significantly large number of videos is published between 9pm and 1am, with another peak at 5am. But remember!this is UTC timezone. 9pm UTC corresponds to 4pm in Washington DC, USA. Therefore, in Washington time, the upload of videos reaches a peak at 4pm, and in the evening from 6 to 8pm. The other peak at 5am UTC corresponds to midnight on the West coast. The worst time to upload is bewteen 3-5am.

## Users are more likely to engage(like, dislike, comment) in which category?
This code for the correlation heatmap was made using the step by step guide from Szabo, B. (2020). How to Create a Seaborn Correlation Heatmap in Python? [online] Medium. Available at: https://medium.com/@szabo.bibor/how-to-create-a-seaborn-correlation-heatmap-in-python-834c0686b88e [Accessed 8 Dec. 2020].

```python
# We create the correlation matrix with .corr
corr_matrix = df.corr()
```

```python
corr_matrix
```

```python
#For greater clarity, we cut the heatmap in half to only have the lower triangle. 
#We use the seaborn heatmap ‘mask’ argument  to cover part of the heatmap.

mask = np.zeros_like(corr_matrix, dtype=np.bool) #takes a boolean array or a dataframe as an argument; when defined, cells become invisible for values where the mask is True
mask[np.triu_indices_from(mask)]= True # isolates the upper triangle of a matrix while turning all the values in the lower triangle into 0
```

```python
f, ax = plt.subplots(figsize=(14, 12)) # Sets the size of the heatmap

heatmap = sns.heatmap(corr_matrix, 
                      mask = mask,
                      square = True,
                      linewidths = .5,
                      cbar_kws = {'shrink': .4, 
                                'ticks' : [-1, -.5, 0, 0.5, 1]},
                      vmin = -1, 
                      vmax = 1, # vmin and vmax set the range of values that serve as the basis for the colormap
                      annot = True, # set to True makes the correlation values become visible on the colored cells
                      annot_kws = {'size': 12})

#add the column names as labels
ax.set_yticklabels(corr_matrix.columns, rotation = 0)
ax.set_xticklabels(corr_matrix.columns)
sns.set_style({'xtick.bottom': True}, {'ytick.left': True})
ax.set_title("Correlation Heatmap for viral videos", fontsize=20) #This line of code sets the title
# The previous line of code was done using Stack overflow (https://stackoverflow.com/questions/32723798/how-do-i-add-a-title-to-seaborn-heatmap)
```
