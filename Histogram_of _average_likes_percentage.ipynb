```python
comedy_filter = df['category'] == 'Comedy'
Animal_filter = df['category'] == 'Animals'
Entertainment_filter = df['category'] == 'Entertainment'
People_filter = df['category'] == 'People'
Music_filter = df['category'] == 'Music'
News_filter = df['category'] == 'News'
Sports_filter = df['category'] == 'Sports'
Film_filter = df['category'] == 'Film'
Autos_filter = df['category'] == 'Autos'
Tech_filter = df['category'] == 'Tech'
Shows_filter = df['category'] == 'Shows'
Travel_filter = df['category'] == 'Travel'
Nonprofit_filter = df['category'] == 'Nonprofit'
Howto_filter = df['category'] == 'Howto'
Education_filter = df['category'] == 'Education'
Games_filter = df['category'] == 'Games'
list_of_filters = [comedy_filter, Animal_filter, Entertainment_filter, People_filter, Music_filter, News_filter, Sports_filter, \
                  Film_filter, Autos_filter, Tech_filter, Shows_filter, Travel_filter, Nonprofit_filter, Education_filter, Games_filter]
```

```python
list_of_aversges = []
for i in list_of_filters:
    path = './viral_meta.csv'
    df = pd.read_csv(path)
    df['percentage_of_likes'] = (((df['num_likes'])/(df['num_dislikes'] + df['num_likes'])) * 100).round()
    try:
        df = df.loc[i, :]
        average = df['percentage_of_likes'].mean()
        list_of_averages += average
    except:
        pass
    
```
