# Investigating the content categories with most trending videos

First of all, we need to replace the category number by the name of the category in our dataset (ie. entertainment, music...
#This matches the category name to the category-id, to be found in the US_category_id.json file (downloaded from https://www.kaggle.com/datasnaek/youtube-new)

NOTE: The following lines of code have been adapted from code by  

```python
# This creates a dictionary to hold a json file which has the info of categories
category_info={}
# by the help of this function we read the json file and then extract the data that we want to the dictionary
with open('.ipynb_checkpoints/US_category_id.json', 'r') as f:
    data = json.load(f)
    for category in data['items']:
        category_info[category["id"]]=category["snippet"]["title"]

print(category_info)
```

```python
# To map category_id and category infos I'm converting category_id to string type
df['category_id']=df['category_id'].astype(str)
# I'm creating Category columns thanks to map func.
df['Category']=df['category_id'].map(category_info)
df['Category'].head()
```

```python
# Counting the video category names
plt.figure(figsize = (14,9))
gr = sns.countplot('Category', data=df, palette="bright", order = df['Category'].value_counts().index)
gr.set_xticklabels(gr.get_xticklabels(),rotation=45)
gr.set_title("Number of trending videos by category", fontsize=20)
gr.set_xlabel("", fontsize=12)
gr.set_ylabel("Count", fontsize=12)
```

## Title length 

This plot was coded by reworking code from Ammar Alyousfi's analysis (2018), to be found at https://notes.quantecon.org/submission/5bca753e95c6070015fa65ff) and Tanmayee W, 2018 from her github repository https://github.com/tanmayeewaghmare/Exploratory-Data-Analysis/blob/master/Youtube%20videos%20analysis_%20US%20Market.ipynb 
Their code served as a basis for the plot that was then completed through trial and error, with help from Stack Overflow. 

```python
df['title_length']=df['title'].apply(lambda x: len(str(x).split()))
```

```python
plt.figure(figsize = (14,20))

plt.subplot(421)
g1 = sns.distplot(df['title_length'],kde=False, rug=True,
                 hist = True)
g1.set_title('Title length of a trending video', fontsize = 14)
g1.set(xlabel='Title Length (in Number of words)', ylabel="Number of videos")
```

We can see that the title-length distribution resembles a normal distribution, where most videos have title lengths between approximately 5 and 13 words.

```python
#Original code by Ammar Alyousfi
PLOT_COLORS = ["#268bd2", "#0052CC", "#FF5722", "#b58900", "#003f5c"]
df["title_length"] = df["title"].apply(lambda x: len(x))

fig, ax = plt.subplots()
_ = sns.distplot(df["title_length"], kde=False, rug=False, 
                 color=PLOT_COLORS[1], hist_kws={'alpha': 1}, ax=ax)
_ = ax.set(xlabel="Title Length", ylabel="No. of videos", xticks=range(0, 110, 10))
```

## Title length and views

We are interested in a scatter plot to find out the relationship between title length and number of views. We adapt 

```python
plt.figure(figsize = (14,20))

fig, ax = plt.subplots()
_ = ax.scatter(y=df['views'], x=df['title_length'], color=PLOT_COLORS[2],edgecolors="#000000", linewidths=0.3)
_ = ax.set(xlabel="Title Length (in number of words)", ylabel="Number of Views")

```

Why did we choose this visualisation? We wanted to explore if there is a relationship between the title length (number of words in title) and the number of views a video can receive, since titles are the most important factor to make a video eye-catching. Does the length of title have an effect on the views? How many words should YouTubers insert in their titles? 

We used the following scatterplot to verify these questions.We found that videos with around 5 to 10 words attract the most views, which implies that YouTubers should consider utilizing short and concised titles when naming their videos. A long title could actually play a negative factor on their number of views.
